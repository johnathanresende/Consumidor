```python
import pandas as pd
pd.set_option('display.max_colwidth', -1) #tamanho maximo das colunas
```

# Grupo 1

```python
df1 = pd.read_table('Resumo_Anais_2012-1-500_tratado.txt',  encoding = "ISO-8859-1")
```

```python
df1
```

```python
df1 = pd.DataFrame(df1.COLUNAS.str.split(':', 1).tolist(), columns = ['Variável', 'Dado']) #colon separated split
```

```python
df1
```

```python
#substituição None Values
df1.replace(to_replace=[None], value='TituloTrabalho', inplace=True)
```

```python
df1
```

```python
df1.to_csv ('group1_splited_python.csv', index = False, header=True)
```
