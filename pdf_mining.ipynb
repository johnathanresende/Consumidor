# PyPDF2
documentação: https://pythonhosted.org/PyPDF2/

```python
# importação da biblioteca
import PyPDF2 as p2 
```

```python
# criar uma variável para armazenar o PDF
arquivoPdf = open ("2015_MariaIvoneteGomesdoNascimento.pdf", "rb")
```

```python
# ler o PDF com a biblioteca PyPDF2
pdfReader = p2.PdfFileReader(arquivoPdf)
```

```python
# obter a primeira página do arquivo
primeira_pag = pdfReader.getPage(0)
```

```python
#extrair o texto da página
print(primeira_pag.extractText())
```

```python
# obter os metadados contidos no documento
print(pdfReader.getDocumentInfo())
```

# pdfminer3

documentação: https://github.com/gwk/pdfminer3/blob/master/README.md

```python
!pip install pdfminer3
```

```python
import pdfminer3
```

```python
#função para obter todo o pdf
from pdfminer3.layout import LAParams, LTTextBox
from pdfminer3.pdfpage import PDFPage
from pdfminer3.pdfinterp import PDFResourceManager
from pdfminer3.pdfinterp import PDFPageInterpreter
from pdfminer3.converter import PDFPageAggregator
from pdfminer3.converter import TextConverter
import io

resource_manager = PDFResourceManager()
fake_file_handle = io.StringIO()
converter = TextConverter(resource_manager, fake_file_handle, laparams=LAParams())
page_interpreter = PDFPageInterpreter(resource_manager, converter)

with open('2015_MariaIvoneteGomesdoNascimento.pdf', 'rb') as fh:

    for page in PDFPage.get_pages(fh,
                                  caching=True,
                                  check_extractable=True):
        page_interpreter.process_page(page)

    text = fake_file_handle.getvalue()

# close open handles
converter.close()
fake_file_handle.close()

print(text)
```
